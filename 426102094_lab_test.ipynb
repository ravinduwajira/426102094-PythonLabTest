```python
import pandas as pd
import kagglehub
from kagglehub import KaggleDatasetAdapter

# ---------------------------------------------------------
# Task 2: Read datasets into Python
# ---------------------------------------------------------

def load_kaggle_dataset(repo_handle, file_name):
    """
    Downloads and loads a Kaggle dataset directly into a Pandas DataFrame.
    """
    try:
        print(f"Downloading and loading: {repo_handle} -> {file_name}")
        df = kagglehub.load_dataset(
            KaggleDatasetAdapter.PANDAS,
            repo_handle,
            file_name
        )
        print("Successfully loaded. First 5 records:")
        print(df.head(), "\n")
        return df
    except Exception as e:
        print(f"Error loading {repo_handle}: {e}")
        return None

# Load Twitter Edge Nodes Dataset
twitter_handle = "mathurinache/twitter-edge-nodes"
twitter_file = "Twitter-dataset/data/edges.csv"
twitter_df = load_kaggle_dataset(twitter_handle, twitter_file)

# Load GitHub Social Network Graph Dataset
github_handle = "gitanjali1425/github-social-network-graph-dataset"
github_file = "musae_git_edges.csv"
github_df = load_kaggle_dataset(github_handle, github_file)

# ---------------------------------------------------------
# Identify Numerical Attributes
# ---------------------------------------------------------

def identify_numerical_attributes(df, dataset_name):
    """
    Identifies and prints the numerical columns in a given DataFrame.
    """
    if df is not None:
        numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
        print(f"Numerical attributes in {dataset_name}: {numerical_cols}")
    else:
        print(f"Cannot identify attributes: {dataset_name} DataFrame is empty.")

identify_numerical_attributes(twitter_df, "Twitter Dataset")
identify_numerical_attributes(github_df, "GitHub Dataset")
```

```python
# ---------------------------------------------------------
# Task 3: Compute summary statistics
# ---------------------------------------------------------

def compute_summary_statistics(df, dataset_name):
    """
    Computes mean, min, max, std, and degree distributions
    for the loaded network datasets.
    """
    if df is None or df.empty:
        print(f"Skipping {dataset_name}: DataFrame is empty or not loaded.")
        return

    print(f"\n{'='*50}")
    print(f"Summary Statistics for {dataset_name}")
    print(f"{'='*50}")

    # Get numerical columns
    numerical_cols = df.select_dtypes(include=['number']).columns.tolist()

    # Calculate basic required statistics
    for col in numerical_cols:
        print(f"\n--- Statistics for column: '{col}' ---")
        print(f"Mean:               {df[col].mean():.2f}")
        print(f"Minimum value:      {df[col].min()}")
        print(f"Maximum value:      {df[col].max()}")
        print(f"Standard deviation: {df[col].std():.2f}")

    # "Any other useful statistical measures" - Node Degree Calculation
    print(f"\n--- Advanced Metric: Node Degree Statistics ---")
    try:
        # Network datasets usually have Source and Target as the first two columns
        source_col = df.columns[0]
        target_col = df.columns[1]

        # Memory-efficient way to count occurrences (degrees) in massive datasets
        source_counts = df[source_col].value_counts()
        target_counts = df[target_col].value_counts()

        # Combine the counts to get total connections per node
        node_degrees = source_counts.add(target_counts, fill_value=0)

        print(f"Total Unique Nodes: {len(node_degrees)}")
        print(f"Average Degree (connections per node): {node_degrees.mean():.2f}")
        print(f"Maximum Degree (most connected 'super-node'): {node_degrees.max()}")
        print(f"Minimum Degree: {node_degrees.min()}")

    except Exception as e:
        print(f"Could not compute node degrees: {e}")

# Run the function for both datasets
compute_summary_statistics(twitter_df, "Twitter Dataset")
compute_summary_statistics(github_df, "GitHub Dataset")
```
